```python
import sys
```

```python
for test_string in ["I love Richard Feynman", "Physics is life", "Life of Pi", "The mathematics of reality"]:
    print(f"### test_string: {test_string} ###")
    for encoding in ["utf-8", "utf-16", "utf-32"]:
        print(f"{encoding} encoding length: {len(test_string.encode(encoding))} encoding: {test_string.encode(encoding)}")
    print("################")
```

```python
def decode_utf8_bytes_to_str_wrong(bytestring: bytes):
    return "".join([bytes([b]).decode("utf-8") for b in bytestring])
```

```python
decode_utf8_bytes_to_str_wrong('Niño'.encode())
```
